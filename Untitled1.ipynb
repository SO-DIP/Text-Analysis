```python
import pyphen
import pandas as pd
import re
from nltk.tokenize import sent_tokenize, word_tokenize


df = pd.DataFrame(columns=[
    "Filename", "Positivity Score", "Negativity Score", "Polarity Score",
    "Subjectivity Score", "Avg Sentence Length", "Percentage of Complex Words",
    "Fog Index", "Avg Number of Words Per Sentence", "Complex Word Count",
    "Word Count", "Avg Sylleble Per Word", "Personal Pronouns", "Avg Word Length"
])

for i in range(0, 114):
    
    with open(r"C:\Text Analysis\url_{}.txt".format(i), 'r', encoding='utf-8') as file:
        # Read the contents of the file
        text_data = file.read()
        
    relevant_text = text_data.lower()
    tokens = word_tokenize(relevant_text)
    tokens = [re.sub(r'[^a-zA-Z0-9]', '', token) for token in tokens if re.sub(r'[^a-zA-Z0-9]', '', token)]
    number_of_words = len(tokens)
    tokens = [token for token in tokens if token not in stop_words]

    positive_score = 0
    negative_score = 0

    # Calculate positive and negative scores
    for word in tokens:
        if word in positive_words:
            positive_score += 1
        if word in negative_words:
            negative_score += 1

    # Calculate polarity score
    polarity_score = (positive_score - negative_score) / (positive_score + negative_score + 0.000001)

    # Calculate subjectivity score
    subjectivity_score = (positive_score + negative_score) / (len(tokens) + 0.000001)

    # Tokenize the text into sentences
    sentences = sent_tokenize(relevant_text)

    # Calculate the average number of words per sentence
    total_words = 0
    for sentence in sentences:
        words = word_tokenize(sentence)
        total_words += len(words)

    average_words_per_sentence = total_words / len(sentences)

    # Calculate the average sentence length
    total_characters = len(relevant_text) - relevant_text.count(' ')
    average_sentence_length = total_characters / len(sentences)

    # Initialize the syllable counter
    dic = pyphen.Pyphen(lang='en')

    # Count the number of complex words
    number_of_complex_words = 0
    for word in tokens:
        syllables = dic.inserted(word).count('-') + 1
        if syllables > 2:
            number_of_complex_words += 1

    # Calculate the percentage of complex words
    percentage_of_complex_words = (number_of_complex_words / len(tokens)) * 100

    fog_index = 0.4 * (average_sentence_length + percentage_of_complex_words)
    total_characters = sum(len(word) for word in tokens)

    # Calculate the total number of words
    total_words = len(tokens)

    # Calculate the average word length
    average_word_length = total_characters / total_words

    word_count = len(tokens)
    syllables_per_word = [dic.inserted(word).count('-') + 1 for word in tokens]

    # Calculate the average syllables per word
    average_syllables_per_word = sum(syllables_per_word) / len(tokens)

    # Define the personal pronouns pattern
    pattern = r"\b(I|we|my|ours|us)\b"

    # Find the personal pronouns in the text
    matches = re.findall(pattern, relevant_text, flags=re.IGNORECASE)

    # Filter out the pronouns that match "US" as a country
    filtered_matches = [match for match in matches if match.lower() != "us"]

    # Count the number of personal pronouns
    count = len(filtered_matches)

    result = {
        "Filename": r"C:\Text Analysis\url_{}.txt".format(i),
        "Positivity Score": positive_score,
        "Negativity Score": negative_score,
        "Polarity Score": polarity_score,
        "Subjectivity Score": subjectivity_score,
        "Avg Sentence Length": average_sentence_length,
        "Percentage of Complex Words": percentage_of_complex_words,
        "Fog Index": fog_index,
        "Avg Number of Words Per Sentence": average_words_per_sentence,
        "Complex Word Count": number_of_complex_words,
        "Word Count": word_count,
        "Avg Sylleble Per Word": average_syllables_per_word,
        "Personal Pronouns": count,
        "Avg Word Length": average_word_length
    }
    
    df = df.append(result, ignore_index=True)
    print(df)
```

```python

output_file_path = r"C:\Text Analysis\out.xlsx"

df.to_excel(output_file_path, index=False)

print("Excel file saved successfully.")
```
